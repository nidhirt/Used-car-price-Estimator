### Ford file

```python
# importing the dataset
ford_df1 = pd.read_csv('data/Ford.csv')
```

```python
ford_df1.head()
```

```python
ford_df1["S"] = ford_df1["S"].replace(1, "S")
ford_df1["SES"] = ford_df1["SES"].replace(1, "SES")
ford_df1["SEL"] = ford_df1["SEL"].replace(1, "SEL")
ford_df1["ST"] = ford_df1["ST"].replace(1, "ST")
ford_df1["Sedan"] = ford_df1["Sedan"].replace(1, "Sedan")
ford_df1["NEW LISTING"] = ford_df1["NEW LISTING"].replace(1, "New Listing")
ford_df1.head()
```

```python
ford_df1.info()
ford_df1["Model_Year"].value_counts()
```

```python
ford_df1["S"] = ford_df1["S"].replace(0, "")
ford_df1["SES"] = ford_df1["SES"].replace(0, "")
ford_df1["SEL"] = ford_df1["SEL"].replace(0, "")
ford_df1["ST"] = ford_df1["ST"].replace(0, "")
ford_df1["Sedan"] = ford_df1["Sedan"].replace(0, "")
ford_df1["NEW LISTING"] = ford_df1["NEW LISTING"].replace(0, "")
ford_df1.head()
```

```python
ford_df1["Model"]=ford_df1["S"].map(str)+ford_df1["SEL"].map(str)+ford_df1["SES"].map(str)+ford_df1["ST"].map(str)+ford_df1["Sedan"].map(str)+ford_df1["NEW LISTING"].map(str)
ford_df1.head()
```

```python
ford_df1["Model"].value_counts()
```

```python
ford_df1["Model"] = ford_df1["Model"].replace("SSedan", "S")
ford_df1["Model"] = ford_df1["Model"].replace("SedanNew Listing", "Sedan")
ford_df1["Model"] = ford_df1["Model"].replace("SELSedan", "SEL")
ford_df1["Model"] = ford_df1["Model"].replace("nannannannannannan", "")
ford_df1["Model"] = ford_df1["Model"].replace("SESNew Listing", "SES")
ford_df1["Model"] = ford_df1["Model"].replace("SELSedanNew Listing", "SEL")
ford_df1["Model"].value_counts()
```

```python
ford_df1["Model_Year"].value_counts()
```

```python
ford_df1clean=ford_df1[["Price","Mileage","Model_Year","Model"]]
```
